```python
def mimic_replace(my_str,sub_str,new_str):
    output=''
    i=0
    while i<len(my_str):
        if my_str[i:i+len(sub_str)] != sub_str:
            output +='*'
            i+=1
        else:
            output +=new_str
            i=i+len(sub_str)
    print(output)
```
