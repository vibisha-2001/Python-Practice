```python
lst=[(5,4),(5,6),(2,3),(3,8)]
for i in range(len(lst)):
    for j in range(len(lst)-i-1):
        if lst[j][-1]>lst[j+1][-1]:
            lst[j],lst[j+1]=lst[j+1],lst[j]
print(lst)
```

```python
import random
```

```python

def mimic_shuffle(lst):
    new_lst=[]
    used=set()
    while len(new_lst)<len(lst):
        i=random.randint(0,len(lst)-1)
        if i not in used:
            used.add(i)
            new_lst.append(lst[i])
    print(new_lst)
    
```

```python
mimic_shuffle([45,32,67,40,20,60])
```
