```python
def mimic_capitalize(s):
    if len(s)==0:
        return s
    else:
        result=''
        fc=s[0]
        if 'a'<=fc<='z':
            result +=chr(ord(fc)-32)
        else:
            result +=fc
        for i in s[1:]:
            if 'A'<=i<='Z':
                result +=chr(ord(i)+32)
            else:
                result +=i
    return result
                                           
```

```python
my_str="hi mY name Is ViBisha"
mimic_capitalize(my_str)
```
