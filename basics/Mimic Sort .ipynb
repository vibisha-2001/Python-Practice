```python
def mimic_sort_asc(lst):
    for i in range(len(lst)):
        for j in range(len(lst)-i-1):
            if lst[j]>lst[j+1]:
                lst[j],lst[j+1] = lst[j+1],lst[j]
    print(lst)
```

```python
mimic_sort_asc([3,2,6,9,4,3,10])
```

```python
def mimic_sort_desc(lst):
    for i in range(len(lst)):
        for j in range(len(lst)-i-1):
            if lst[j]<lst[j+1]:
                lst[j],lst[j+1] = lst[j+1],lst[j]
    print(lst)
```

```python
mimic_sort_desc([1,4,2,9,10,3])
```
